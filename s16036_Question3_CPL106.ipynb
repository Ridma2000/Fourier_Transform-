```python
# Question 3
# Part II

# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft, ifft

# Defining wave properties
f1 = 100  # Frequency of wave 1 in Hz
f2 = 200  # Frequency of wave 2 in Hz
A1 = 2    # Amplitude of wave 1 in Volts
A2 = 1    # Amplitude of wave 2 in Volts
sr = 1000  # Sampling rate in samples per second (Hz)
ts = 1/sr  # Sampling interval (time between samples)
duration = 1  # Duration of the signal in seconds

# Generating the time array 't' using linspace
# 'endpoint=False' ensures that the last point in the time array is not included
# This results in a time array that ranges from 0 to slightly below 'duration'
t = np.linspace(0, duration, int(sr * duration), endpoint=False)

# Generating two sine waves with the given frequencies and amplitudes
y1 = A1 * np.sin(2 * np.pi * f1 * t)  # Sine wave 1 with frequency f1
y2 = A2 * np.sin(2 * np.pi * f2 * t)  # Sine wave 2 with frequency f2

# Superposing (adding) the two sine waves to get the resultant wave
result = y1 + y2

# Performing Fast Fourier Transform (FFT) on the resultant wave
# FFT converts the time-domain signal into its frequency components
fft_result = np.fft.fft(result) #computes the Fast Fourier Transform (FFT) of the result signal

# Generating the corresponding frequency array
# 'fftfreq' returns the frequencies corresponding to the FFT result
f = np.fft.fftfreq(len(result), 1/sr)

# Plotting the magnitude of the FFT result against the frequency
plt.plot(f, np.abs(fft_result))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (V)')
plt.title('Amplitude vs Frequency')
plt.xlim(0, 450)  # Limiting the x-axis to show frequencies up to 450 Hz
plt.grid(True)
plt.show()
```

```python
#part III
#(d)
import numpy as np
import matplotlib.pyplot as plt

fc = 500 #frequency of carrier wave
fm = 50 #frequency of modulating wave
Ac = 5 # Amplitude of carrier wave
Am = 2 # Amplitude of modulating wave
end = 0.1 #end of range
sr = 1000     #sampling rate
t = np.linspace(0, end, sr) 

yc = Ac * np.sin(2*np.pi*fc*t)   #carrier wave
ym = Am * np.sin(2*np.pi*fm*t)
yt = (Ac + ym) * np.sin(2*np.pi*fc*t) 

plt.figure(figsize=(9, 5))
plt.subplot(3, 1, 1)
plt.plot(t, yc)
plt.xlabel('Time')
plt.ylabel('yc')
plt.title('Carrier wave vs time')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, ym)
plt.xlabel('Time')
plt.ylabel('ym')
plt.title('Modulating signal vs time')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, yt)
plt.xlabel('Time')
plt.ylabel('yt')
plt.title('Magnitude of the signal vs time')
plt.grid(True)

plt.tight_layout()   #aligns the graphs one below the other
plt.show()

```

```python
#(e)
print("(e)")
print("i.")
# Perform FFT on the modulated signal
f_components = np.fft.fft(yt) # frequency components
f = np.fft.fftfreq(len(t), t[1] - t[0]) # generates the frequency bins corresponding to the Fast Fourier Transform (FFT) output
#t[1] - t[0] is the time gap

plt.figure(figsize=(9, 6))
plt.plot(f, np.abs(f_components))
plt.xlabel('Frequency  (Hz)')
plt.ylabel('Amplitude (V)')
plt.title('Fourier spectrum of modulated signal')
plt.xlim(0, 2000)  # set the limits of the x-axis on a plot to the range from 0 to 2000
plt.grid(True)
plt.show()

print("ii.")
#finding lowest frequency
lowest_frequency = np.abs(f[np.argmin(np.abs(f_components))])
#np.argmin() returns the index of the smallest value in the array passed to it.

print("Lowest frequency = ",lowest_frequency, "Hz")
print()

print("iii.")
#finding highest frequency
highest_frequency = np.abs(f[np.argmax(np.abs(f_components))])
#np.argmax() returns the index of the Highest value in the array passed to it.

print("Highest frequency = ",highest_frequency, "Hz")
print()

print("iv.")
#bandwith value
print("Bandwidth = ", highest_frequency - lowest_frequency , "Hz")
```

```python
#part IV
print("IV")
print("b)")
import numpy as np
import matplotlib.pyplot as plt

sr = 1000 # sampling rate
t = np.linspace(0,0.1,sr)

ycc = 20 * np.sin(400*np.pi*t)
ymm = 10 * np.sin(200*np.pi*t) + 40 * np.sin(100*np.pi*t)
ytt = (20 + ymm) * np.sin(400*np.pi*t)

plt.figure(figsize=(9, 5))
plt.subplot(3, 1, 1)
plt.plot(t, ycc)
plt.xlabel('Time')
plt.ylabel('yc')
plt.title('Carrier wave vs time')
plt.grid(True)


plt.subplot(3, 1, 2)
plt.plot(t, ymm)
plt.xlabel('Time')
plt.ylabel('ym')
plt.title('Modulating signal vs time')
plt.grid(True)


plt.subplot(3, 1, 3)
plt.plot(t, ytt)
plt.xlabel('Time')
plt.ylabel('yt')
plt.title('Total signal vs time')
plt.grid(True)

plt.tight_layout()
plt.show()
```
