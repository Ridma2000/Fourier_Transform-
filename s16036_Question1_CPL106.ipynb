```python
# Question 1

# Importing required libraries
import numpy as np  # For numerical operations, particularly with arrays
import matplotlib.pyplot as plt  # For plotting graphs

# Function to calculate y(t), representing a sinusoidal wave
def y(t):
    A = 10  # Amplitude of the sine wave
    F = 50  # Frequency of the sine wave in Hz
    phi = np.pi / 2  # Phase shift of the sine wave in radians
    return A * np.sin((2 * np.pi * F * t) + phi)  # Returns the sine wave formula

# Function to create an array of time values from 0 to 10 with a step size t
def x(t):
    return np.arange(0, 10, t)  # Returns an array of time values from 0 to 10

# Generating y(t) values for different step sizes of time
y1 = [y(t) for t in x(0.1)]    # y-values for the first time step (10Hz)
y2 = [y(t) for t in x(0.01)]   # y-values for the second time step (100Hz)
y3 = [y(t) for t in x(0.001)]  # y-values for the third time step (1000Hz)
y4 = [y(t) for t in x(0.0001)] # y-values for the fourth time step (10000Hz)

# Plotting the sine waves for different frequencies
plt.figure(figsize=(12, 8))  # Creates a new figure with a specified size
plt.plot(x(0.1), y1, label='10Hz', color='purple')  # Plots the sine wave with a 10Hz frequency
plt.plot(x(0.01), y2, label='100Hz', color='green')  # Plots the sine wave with a 100Hz frequency
plt.plot(x(0.001), y3, label='1000Hz', color='black')  # Plots the sine wave with a 1000Hz frequency
plt.plot(x(0.0001), y4, label='10000Hz', color='red')  # Plots the sine wave with a 10000Hz frequency

# Setting the x-axis limits for better visualization
plt.xlim(0, 0.04)  # Limits the x-axis from 0 to 0.04 seconds

# Adding labels and title to the plot
plt.xlabel('Time (s)')  # Label for the x-axis
plt.ylabel('Displacement (m)')  # Label for the y-axis
plt.title('Sine waves')  # Title of the plot

# Adding a legend to distinguish between the different sine waves
plt.legend()  # Displays the legend

# Displaying the plot
plt.show()  # Shows the plot with the sine waves
```
